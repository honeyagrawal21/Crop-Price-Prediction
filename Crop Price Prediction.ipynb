```python
def detect_outliers(data2,features):
    outlier_indices = []
    for col in features:
        Q1 = np.percentile(data2[col], 25)
        Q3 = np.percentile(data2[col],75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR
        outlier_list_col = data2[(data2[col] < Q1 - outlier_step) | (data2[col] > Q3 + outlier_step )].index
        outlier_indices.extend(outlier_list_col)  

    return outlier_indices
```

```python
len(detect_outliers(data2,["modal_price"]))
```

```python
#IQR
Q1=np.percentile(data2["modal_price"],25,interpolation="midpoint")
Q3=np.percentile(data2["modal_price"],75,interpolation="midpoint")
IQR=Q3-Q1
```

```python
upper=np.where(data2["modal_price"]>=(Q3+1.5*IQR))
lower=np.where(data2["modal_price"]<=(Q1-1.5*IQR))
```
